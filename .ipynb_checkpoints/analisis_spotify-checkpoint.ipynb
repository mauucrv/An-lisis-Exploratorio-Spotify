```python
# Importar las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos
df = pd.read_csv('spotify-2023.csv', encoding='latin-1')
```

```python
# Visualizar datos rápidamente
print("Primeras 5 filas del dataset:")
print(df.head())
print("\nInformación general y tipos de datos:")
df.info()
```

```python
# 10 artistas con más canciones en el Top
top_10_artistas = df['artist(s)_name'].value_counts().head(10)
```

```python
# Gráfico de barras para representar a los 10 artistas con más canciones en el top
plt.figure(figsize=(10, 6))
sns.barplot(x = top_10_artistas.values, y = top_10_artistas.index, hue =  top_10_artistas.index, palette= 'viridis', legend = False)
plt.title('Top 10 Artistas con Más Canciones en el Ranking')
plt.xlabel('Número de Canciones')
plt.ylabel('Artista')
plt.show()
```

```python
# Histograma para responder: ¿Cómo se distribuye la "bailabilidad" (danceability) de las canciones?
# Danceability: 0 = poco bailable, 100 = muy bailable

plt.figure(figsize=(10, 6))
sns.histplot(df['danceability_%'], bins=20, kde=True, color='purple')
plt.title('Distribución de la Bailabilidad de las Canciones')
plt.xlabel('Porcentaje de Bailabilidad (%)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Celda 6: Conclusión simple
print("\nAnálisis completado.")
print(f"El artista con más canciones en el top es: {top_10_artistas.index[0]}")
```
